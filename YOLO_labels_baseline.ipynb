```python
from ultralytics import YOLO
import cv2
import glob
import os
# Load the YOLOv8 model
model = YOLO('yolov8n.pt')

# Initialize cumulative vehicle counts

test_accident_path='/home/nilanjanac/experiments/llm_project/test/accident/'
test_nonaccident_path='/home/nilanjanac/experiments/llm_project/test/non_accident/'
video_paths_accident=[v for v in os.listdir(test_accident_path)]
video_paths_non_accident=[v for v in os.listdir(test_nonaccident_path)]
# Path to your video frames
a={}
for v in video_paths_non_accident:
    total_vehicle_count = {
    'car': 0,
    'truck': 0,
    'bike': 0,
    'bicycle': 0
}
    frame_paths = glob.glob(f'{test_nonaccident_path}{v}/*.jpg')  # Replace with the correct path to your frames
    #frame_paths = glob.glob(f'{test_non_accident_path}{v}/images/*.jpg')  
    # Loop over each frame
    for frame_path in frame_paths:
        # Read each frame (image)
        frame = cv2.imread(frame_path)

        # Perform YOLO inference on the current frame
        results = model(frame)

        # Count vehicles in the current frame
        vehicle_count = {
            'car': 0,
            'truck': 0,
            'bike': 0,
            'bicycle': 0
        }

        for result in results:
            for box in result.boxes:
                class_name = model.names[int(box.cls)]  # Get class name
                
                if class_name == 'car':
                    vehicle_count['car'] += 1
                elif class_name == 'truck':
                    vehicle_count['truck'] += 1
                elif class_name == 'motorbike':  # 'bike' in YOLO models refers to motorbike
                    vehicle_count['bike'] += 1
                elif class_name == 'bicycle':
                    vehicle_count['bicycle'] += 1

        # Aggregate vehicle counts across all frames
        total_vehicle_count['car'] = max(vehicle_count['car'],total_vehicle_count['car'])
        total_vehicle_count['truck'] = max(vehicle_count['truck'],total_vehicle_count['truck'])
        total_vehicle_count['bike'] = max(vehicle_count['bike'],total_vehicle_count['bike'])
        total_vehicle_count['bicycle'] = max(vehicle_count['bicycle'],total_vehicle_count['bicycle'])
    print(f"Total vehicles detected in video: {total_vehicle_count}")
    a[v]=total_vehicle_count
```

```python
import json
with open('/home/nilanjanac/experiments/llm_project/test/non_accident_vehicle_count_YOLO.json', 'w') as json_file:
    json.dump(a, json_file, indent=4)
```
